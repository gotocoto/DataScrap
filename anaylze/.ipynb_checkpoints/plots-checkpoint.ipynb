```python
#Graph the comments per hour per day of the week
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import json
# Connect to MySQL
with open('../db_config.json', 'r') as config_file:
    config = json.load(config_file)

conn = mysql.connector.connect(**config)

# Query to executes
query = """
SELECT
    HOUR(time) AS hour,
    DAYOFWEEK(time) AS day_of_week,
    COUNT(*) AS count
FROM comment 
GROUP BY hour, day_of_week
ORDER BY hour, day_of_week;
"""

# Load data into a DataFrame
df = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Prepare data for plotting
# Pivot data to have hours as rows and days of the week as columns
df_pivot = df.pivot_table(index='hour', columns='day_of_week', values='count', fill_value=0)

# Convert comment counts to millions
df_pivot_millions = df_pivot / 1_000_000

# Define day names for better readability in the plot
day_names = {
    1: 'Sunday', 2: 'Monday', 3: 'Tuesday', 4: 'Wednesday',
    5: 'Thursday', 6: 'Friday', 7: 'Saturday'
}
print("running")
# Plot data
plt.figure(figsize=(14, 9))
for day in df_pivot_millions.columns:
    plt.plot(df_pivot_millions.index, df_pivot_millions[day], label=day_names[day])

plt.xlabel('Hour of the Day', fontsize=14)
plt.ylabel('Number of Comments (in Millions)', fontsize=14)
plt.title('Comments per Hour for Each Day of the Week', fontsize=16)
plt.xticks(range(0, 24), [f'{h}' for h in range(24)], fontsize=12)  # Format x-axis as hours with ":00"
plt.yticks(fontsize=12)
plt.legend(title='Day of the Week', title_fontsize='13', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import json
import re

# Regular expression to clean HTML tags, links, and non-alphabetical characters
clean_re = re.compile(r'<.*?>|https?://\S+|[^a-zA-Z\s]')

# Load database configuration
with open('../db_config.json', 'r') as config_file:
    config = json.load(config_file)

# Connect to MySQL
conn = mysql.connector.connect(**config)

# Query to fetch comment content and upvotes (limiting to 1 million comments)
query = """
SELECT content, ranks_up
FROM comment
LIMIT 10000000;
"""

# Load data into a DataFrame
df_content_ranks_up = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Function to clean HTML tags and non-alphabetical characters from comments
def clean_comment(comment):
    return comment
    if pd.isna(comment):
        return ''
    return clean_re.sub('', comment).strip()

# Clean comments and calculate their length
df_content_ranks_up['clean_comment'] = df_content_ranks_up['content'].apply(clean_comment)
df_content_ranks_up['comment_length'] = df_content_ranks_up['clean_comment'].apply(len)

# Create scatter plot
plt.figure(figsize=(14, 9))
plt.scatter(df_content_ranks_up['comment_length'], df_content_ranks_up['ranks_up'], alpha=0.5, c='blue', edgecolors='w', s=10)
plt.xlabel('Length of Comment', fontsize=14)
plt.ylabel('Number of Upvotes', fontsize=14)
plt.title('Comment Length vs. Number of Upvotes', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load database configuration
with open('../db_config.json', 'r') as config_file:
    config = json.load(config_file)

# Connect to MySQL
conn = mysql.connector.connect(**config)

# SQL query with basic HTML tag removal and length calculation
query = """
SELECT LENGTH(REGEXP_REPLACE(content, '<.*?>|https?://\\S+|[^a-zA-Z\\s]', '')) AS cleaned_length, ranks_up
FROM comment
LIMIT 1000000;
"""

# Load data into a DataFrame
df = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Create scatter plot
plt.figure(figsize=(14, 9))
plt.scatter(df['cleaned_length'], df['ranks_up'], alpha=0.5, c='blue', edgecolors='w', s=10)
plt.xlabel('Length of Comment', fontsize=14)
plt.ylabel('Number of Upvotes', fontsize=14)
plt.title('Comment Length vs. Number of Upvotes', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load database configuration
with open('../db_config.json', 'r') as config_file:
    config = json.load(config_file)

# Connect to MySQL database
conn = mysql.connector.connect(**config)

# SQL query to get comment counts per day
query = """
SELECT DATE(time) AS date, COUNT(*) AS comment_count
FROM comment
GROUP BY date
ORDER BY date;
"""

# Load data into a DataFrame
df = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(df['date'], df['comment_count'], marker='o', linestyle='-', color='b')

# Formatting
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Comments', fontsize=14)
plt.title('Comment Activity Over Time', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels

# Show plot
plt.show()
```
